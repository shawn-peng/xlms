```python
from scipy.stats import beta, norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
wc_rv = beta(300, 7700)
```

```python
x = np.linspace(0.02, 0.05, 10000)
plt.plot(x, wc_rv.pdf(x))
```

## Normal approximation

```python
wc_rv.__dict__
```

```python
def normal_approx(beta_rv):
    a, b = beta_rv.args
    u = a / (a + b)
    s = np.sqrt((a * b) / ((a+b) ** 2 * (a+b+1)))
    print(u, s)
    return norm(u, s)
wc_nrv = normal_approx(wc_rv)
```

```python
# x = np.linspace(0, 1, 10000)

plt.plot(x, wc_nrv.pdf(x))
plt.plot(x, wc_rv.pdf(x))
```

```python
plt.plot(x, wc_nrv.cdf(x) - wc_rv.cdf(x))
```

```python
xl = np.log(x / (1-x))
plt.plot(xl, wc_rv.pdf(x))

# plt.plot(xl, wc_nrv.pdf(x))
```

```python
plt.plot(xl, wc_nrv.cdf(x) - wc_rv.cdf(x))
```
