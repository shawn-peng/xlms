```python
def plot_psms(res_file):
    rf = open(res_file)
    # res_csv = csv.DictReader(rf, delimiter='\t')
    res_table = pd.read_csv(rf, sep='\t')
    
    top_hits = res_table.groupby('Spectrum').apply(lambda x: x.iloc[0]).sort_values('Rank')
    
    plt.figure()
    top_hits['pr(alpha=T,beta=T)'].hist(density=True, bins=50)
    plt.figure()
    top_hits['pr(alpha=T,beta=T)'].apply(lambda x: np.log(x)).hist(density=True, bins=50)
```
